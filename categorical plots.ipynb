```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='level of education')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='magma')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df, x='level of education', y='salary',estimator =np.mean, ci="sd")
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df, 
            x='level of education', 
            y='salary',
            estimator =np.mean, 
            ci="sd", 
            hue='division')

plt.legend(bbox_to_anchor=(1.05,1))
```
