Scatter plots show relationship between two continous features. e.g. age, height, salary, etc

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12,8),dpi=180)
sns.scatterplot(data = df, x = 'salary', y= 'sales', hue='level of education', palette='tab10')
```

```python
 plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(data = df, x = 'salary', y= 'sales', style='level of education', hue = 'level of education')

p
```
