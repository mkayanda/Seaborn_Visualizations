```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, x='test preparation course', y='math score')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, x='parental level of education', y='math score')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, x='parental level of education', 
            y='math score', 
            hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.boxplot(data=df, y='parental level of education', 
            x='math score', 
            hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.violinplot(data=df, 
               x='parental level of education', 
            y='math score', 
            palette='Set2_r')
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.violinplot(data=df, 
               x='parental level of education', 
            y='math score', 
            palette='Set2_r', 
              inner='quartiles')
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.violinplot(data=df, 
               x='parental level of education', 
            y='math score', 
            palette='Set2_r', 
              inner='stick')
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.violinplot(data=df, 
               x='parental level of education', 
            y='math score', 
            palette='Set2_r', 
              bw=0.2)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data = df, x = 'math score', 
              y = 'gender', size = 2, 
             hue='test preparation course',
             dodge=True)

plt.legend(bbox_to_anchor=(1.35,0.25))
```
