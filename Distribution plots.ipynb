### Rug Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(10,8))
    
sns.rugplot(x='salary', data=df, height = 0.5)
```

## Histogram

```python
#sns.set(style='ticks')
sns.displot(x='salary', data= df, bins=20)
```

```python
#sns.set(style='ticks')
plt.figure(figsize=(10,8))
sns.displot(x='salary', data= df, bins=20, color = 'pink', edgecolor = 'blue')
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df, x='salary')
```

```python
#sns.set(style='ticks')
sns.displot(x='salary', data= df, bins=20, kde=True)
```

```python
sns.kdeplot(data=df, x='salary', clip=[0,160000], bw_adjust=0.3)
```
